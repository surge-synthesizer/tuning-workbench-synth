```python
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
import numpy as np


plt.rcParams["figure.figsize"] = (10,8)
```

```python
x = np.linspace( 0, 1, 100 )
def saw(x):
    
    if( x < 0.94 ):
        res = ( (x+.3)**2 - .3**2 ) / ( 1.4**2 - .3**2 )
    else:
        res = - (x-1) ** 4 * 10000 + 1 
    return res

sv = np.vectorize( saw )
plt.plot( x, sv(x) )
```

```python
x = np.linspace( 0, 1, 100 )
def tri(x):
    
    if( x < 0.5 ):
        res = ( (x+.3)**2 - .3**2 ) / ( .8**2 - .3**2 )
    else:
        res = 1 - ( (x-.2)**2 - .3**2 ) / ( .8**2 - .3**2 )
    return res

tv = np.vectorize( tri )
plt.plot( x, tv(x) )
```

```python
x = np.linspace( 0, 1, 100 )
def sqr(x):
    
    if( x < 0.5 ):
        res = 1 - 0.1 * ( (x+.3)**2 - .3**2 ) / ( .8**2 - .3**2 )
    else:
        res = .1 * ( (x-.2)**2 - .3**2 ) / ( .8**2 - .3**2 )
    return res

qv = np.vectorize( sqr )
plt.plot( x, qv(x) )
```
